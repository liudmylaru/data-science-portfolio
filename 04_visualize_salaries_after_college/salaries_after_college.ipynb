## Visualizing Salaries Based On College Majors

I'll use a dataset on the job outcomes of students who graduated from college between 2010 and 2012. The original data on job outcomes was released by `American Community Survey`, which conducts surveys and aggregates the data. `FiveThirtyEight` cleaned the dataset and released it on their `Github repo` - [link](https://github.com/fivethirtyeight/data/tree/master/college-majors).

Each row in the dataset represents a different major in college and contains information on gender diversity, employment rates, median salaries, and more. Here are some of the columns in the dataset:

* **Rank** - Rank by median earnings (the dataset is ordered by this column).
* **Major_code** - Major code.
* **Major** - Major description.
* **Major_category** - Category of major.
* **Total** - Total number of people with major.
* **Sample_size** - Sample size (unweighted) of full-time.
* **Men** - Male graduates.
* **Women** - Female graduates.
* **ShareWomen** - Women as share of total.
* **Employed** - Number employed.
* **Median** - Median salary of full-time, year-round workers.
* **Low_wage_jobs** - Number in low-wage service jobs.
* **Full_time** - Number employed 35 hours or more.
* **Part_time** - Number employed less than 35 hours.

I would like to vizualize data set to understand:

- Do students in more popular majors make more money? (*using scatter plots*)
- How many majors are predominantly male? Predominantly female? (*using histograms*)
- Which category of majors have the most students? (*using bar plots*)

```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

recent_grads = pd.read_csv("recent-grads.csv")

# the first row of data set
print(recent_grads.iloc[0])
```

```python
# first and last rows in data set
print(recent_grads.head())
print(recent_grads.tail())
```

```python
# generate summary statistics for all of the numeric columns
recent_grads.describe()
```

I will drop rows with missing values. Matplotlib expects that columns of values we pass in have matching lengths and missing values will cause matplotlib to throw errors.

```python
# the number of rows in recent_grads
raw_data_count = recent_grads.count()
raw_data_count
```

```python
# drop rows containing missing values
recent_grads.dropna(axis=0, inplace=True)
cleaned_data_count = recent_grads.count()
cleaned_data_count
```

Result: only one row contained missing values and was dropped.

Use the plots to explore the following questions:

## Do students in more popular majors make more money?

```python
recent_grads.plot(x='Sample_size', y='Median', kind='scatter', 
                  title='Sample_size vs. Median salary of full-time', 
                  fontsize=12, figsize=(5,5))
```

```python
recent_grads.plot(x='Sample_size', y='Unemployment_rate', kind='scatter', 
                  title='Sample_size vs. Unemployment', 
                  fontsize=12, figsize=(5,5))
```

## Is there any link between the number of full-time employees and median salary?

```python
recent_grads.plot(x='Full_time', y='Median', kind='scatter', 
                  title='Full_time vs. Median of salary of full-time', 
                  fontsize=12, figsize=(5,5))
```

## Do students that majored in subjects that were majority female make more money?

```python
recent_grads.plot(x='ShareWomen', y='Unemployment_rate', kind='scatter', 
                  title='ShareWomen vs. Unemployment_rate', 
                  fontsize=12, figsize=(5,5))
```

## Is there difference between median salay and gender?

```python
recent_grads.plot(x='Men', y='Median', kind='scatter', 
                  title='Men vs. Median', 
                  fontsize=12,
                  figsize=(8,5), xlim=(0, 200000), ylim=(20000, 80000))
```

```python
recent_grads.plot(x='Women', y='Median', kind='scatter', 
                  title='Women vs. Median',
                  fontsize=12,
                  figsize=(8,5), xlim=(0, 200000), ylim=(20000, 80000))
```

## Combines both scatter plots and histograms 

Let's create individual scatter plots to visualize potential relationships between columns and histograms to visualize the distributions of individual columns.
